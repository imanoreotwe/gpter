```python
# pytorch simple example
import torch
import torch.nn as nn

torch.manual_seed(1337) 
emb = nn.Embedding(100000, 100000)
print(emb(torch.LongTensor([[1,2,3,4],[4,3,2,1]])))
```

```python
# tinygrad simple example
from tinygrad import Tensor, nn, dtypes
Tensor.manual_seed(1337)

emb = nn.Embedding(100000, 100000)
print(emb(Tensor([[1,2,3,4],[4,3,2,1]], dtype=dtypes.long)).numpy()) # oom
```
